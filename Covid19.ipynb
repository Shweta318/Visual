```python
data_source=covid_rev['DATA_SOURCE_NAME'].value_counts()
```

```python
date=covid_rev[covid_rev['REPORT_DATE'].between("2020","2021")]
```

```python
covid_rev['PEOPLE_POSITIVE_NEW_CASES_COUNT'].fillna(covid_rev['PEOPLE_POSITIVE_NEW_CASES_COUNT'].mean(),inplace=True)
```

```python
covid_rev
```

```python
sns.scatterplot(data=covid_rev,x='PEOPLE_POSITIVE_NEW_CASES_COUNT',y='PEOPLE_DEATH_COUNT',hue='DATA_SOURCE_NAME')
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=covid_rev,x='REPORT_DATE',y='PEOPLE_POSITIVE_NEW_CASES_COUNT',hue='DATA_SOURCE_NAME')
plt.tight_layout()
```
